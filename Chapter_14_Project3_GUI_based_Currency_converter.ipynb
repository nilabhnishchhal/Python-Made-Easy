```python
import requests 
# creating class for currency converter  
class Currency_Converter: 
    # creating empty dict to store the conversion rates 
    rates = {}  
    def __init__(self, url): 
        # REquesting data from the url or webpage
        data = requests.get(url).json() 
  
        # Extracting the rates from the data 
        self.rates = data["rates"]  
  
    # function to do the actual conversion
    def convert(self, from_currency, to_currency, amount): 
        initial_amount = amount
        # as the base currency is EUR, hence this conversion 
        if from_currency != 'EUR' : 
            amount = amount / self.rates[from_currency] 
        output = amount * self.rates[to_currency]
        # Rounding off the conversion at 2 dcimal places
        output = round(amount * self.rates[to_currency], 2)
        # Output in nice print format
        print('{} {} = {} {}'.format(initial_amount, from_currency, output, to_currency)) 
  
 # Driver code 
if __name__ == "__main__": 
  
    # This access key may not work for you. Generate your own access key
    ACCESS_KEY = "1cf94c2fd27ab40a0b510ccec2d0014b" 
    url = str.__add__('http://data.fixer.io/api/latest?access_key=', ACCESS_KEY)   
    # calling a class instance
    converter = Currency_Converter(url)
    # Taking Inputs from the user
    from_country = input("Convert From Currency: ") 
    to_country = input("Convert to Currency: ") 
    amount = int(input("Amount to be Converted: ")) 
  
    converter.convert(from_country, to_country, amount) 
```
