# Chapter - Pandas in Action

```python
import pandas as pd 

titanic = pd.read_csv("titanic.csv", index_col = 0)

titanic.head()
```

```python
titanic["Fare_Pound"] = titanic["Fare"]*0.80

titanic.head()
```

```python
titanic_1 = titanic.rename(columns= {"Ticket":"Ticket Number", "Fare":"Fare_Dollar"})

titanic_1.head()
```

```python
titanic_1 = titanic_1.rename(columns=str.upper)

titanic_1.head()
```

```python
titanic["Age"].mean()
```

```python
titanic[["Age", "Fare"]].median()
```

```python
titanic[["Age", "Fare"]].describe()
```

```python
titanic.agg({'Age': ['min', 'max', 'median', 'skew'], 
             'Fare': ['min', 'max', 'median', 'mean']})
```

```python
titanic.groupby("Sex").count()
```

```python
titanic[["Name", "Sex"]].groupby("Sex").count()
```

```python
titanic[["Sex", "Age"]].groupby("Sex").mean()
```

```python
titanic.groupby("Sex")["Age"].mean()
```

```python
titanic.groupby(["Sex", "Pclass"])["Fare"].mean()
```

```python
titanic.sort_values(by="Age").head()
```

```python
titanic.sort_values(by=["Pclass", "Age"], ascending= False).head()
```

```python
titanic_male = titanic[titanic["Sex"] == "male"]

titanic_female = titanic[titanic["Sex"] == "female"]
```

```python
titanic_male.head()
```

```python
titanic_female.head()
```

```python
print(titanic_female.shape)

print(titanic_male.shape)
```
